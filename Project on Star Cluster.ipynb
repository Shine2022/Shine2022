# Introduction of Omega Centauri

### Omega Centauri is the Milky Way galaxy’s largest and most luminous globular star cluster, AKA NGC 5139. The star cluster can be found in the constellation Centaurus and can be seen by the naked eye to observers at low northern latitudes and in the southern hemisphere. It contains approximately 10 million stars and is about 16,000 light-years away.[1]  150 light-years in diameter and packed with about numerous stars much older than the Sun, Omega Cen consist of stars with the same age and composition, it also exhibits the presence of different stellar populations with a spread of ages and chemical abundances. In fact, Omega Cen may be the remnant core of a small galaxy merging with the Milky Way.[2] In this paper, I will also discuss the metallicity-age correlation, the stellar kinematics of ω Cen.

## 1.1 Location of Omega Centauri

### Historical studies have paid particular attention to the different mean proper motions that have been detected among the three main sub-populations in ω Cen [10]. The incompleteness of data has caused controversy regarding the population’s accreted origin. Researchers attempted to solve the problem by using astrometry from the second Gaia data release and the recent high-quality, multi-band photometry. To contain enough stars in the RGB-a population, the researchers combine the photometry with the proper motions from other’s findings. They restricted the section by adding a median absolute deviation (MAD) of the proper motion and parallax distributions of members, refined after a few iterations, and set to (1.09, 1.29) mas yr−1 and 0.69 mas, respectively[10]. Following the selection, the sample was made up of 5113 stars. And the typical (median) uncertainties on the individual stars are in the range of 0.08–0.12 mas yr−1in the case of µα∗ and 0.12–0.17 mas yr−1 in the case of µδ. They also adjust their dataset for a function of magnitude and color. The conclusion is obvious----that the three sub-populations have compatible mean motions with each other, inferring that the metal-rich sub-population would not the separate accreted origin [10].

## 2.3 True space velocity

```python
Average_pmra = np.mean(X1)
#μ2=μRA^2 + μDec^2
#VT= μD
μ_ = np.sqrt(np.square(Average_pmra)+np.square(Average_pmdec))
print('estimate for proper motion=',μ_)
Average_Tangential_Velocity = (μ_) * Average_distance * 4.74 * 0.001
print('estimate for Tangential Velocity=',Average_Tangential_Velocity)

Estimate_for_space_velocity = np.sqrt(np.square(Average_Tangential_Velocity)+np.square(estimates_for_radial_velocity))
print('Estimate for space velocity=',Estimate_for_space_velocity)
```

### According to the calculations above, the estimate for proper motion of ω Cen is 24.81 milliarcsecond per year, the estimate for tangential velocity is 210.52 km/s, the estimate for radial velocity is 212.15 km/s (move away from us) and the estimate for the true space velocity approximates to 298 km/s. The dispersion decreases outwards, but the central value is the largest ever measured in any globular cluster.[3] In respect to the Sun's orbital velocity, Omega Centauri is moving away at a relative fast speed. The concerns regarding to the estimates are numerous, namely, the error bound of the database and outliers of datas are required to be considered. Rotation of the stars in ω Cen can also induce geometric effects, affecting the effective temperature and luminosity. It is now clear that clusters host stars with a range of rotational velocities, which can have a strong effect on the observed colour-magnitude diagram of the cluster.[6]


## 3.1 The Color-Absolute Magnitude Diagram 

```python
parallax_1, = np.where((gaiarv_cat['parallax'] > 0.07)&(gaiarv_cat['parallax'] < 0.35)\
                        &(gaiarv_cat['ra'] >= 201.1)&(gaiarv_cat['ra'] < 201.8)&(gaiarv_cat['dec'] >= -48)\
                        &(gaiarv_cat['dec'] < -47.2))
                        
distance_1 = 1000.0/(gaiarv_cat['parallax'][parallax_1])
distance_1.name = 'distance(pc)'
absmag = gaiarv_cat['phot_g_mean_mag'][parallax_1] - 5*np.log10(distance_1/10.0)
absmag.name = 'abs_mag_2g'

fig = pyplot.figure(figsize=(15,5))
ax = pyplot.subplot(1,1,1)
ax.plot(gaiarv_cat['bp_rp'][parallax_1],absmag,'r,',alpha=1)
ax.set_xlabel('BP-RP')
ax.set_ylabel('$M_G$')
ax.set_xlim(-5,10)
ax.set_ylim(10,-15)
```

The reason why it is so luminous partially due to its mass. ### The underlying correlation between the metallicity of ω Cen and its estimated age follows the patterns for most astronomical objects in the Milky Way Galaxy. The halo stars in the Milky Way Galaxy have average metallicities of <[Fe / H]> = -1.5 dex and a maximum metallicity of ~ -1.0 dex, where ω Cen located at[7]. In early studies, scientists have found that ω Cen showed star-to-star abundance variations of the iron-peak elements, unlike other Galactic Globular clusters, of which the majorities exhibit large variations in the lighter elements such as CNO, Na, Mg, and Al. The red-giant branch (RGB) is the portion of the red giants whose helium ignition occurs in the course of stellar evolution[8]. They developed a shape of isochrones to divide the observed RGB stars into four subpopulations directly in the CMD, assuming the shape of the cluster RGB with different metallicity can be adequately represented by isochrones.[9] The Y^2 Isochrones they used, had taken into consideration of Helium diffusion and convective core overshoot. They obtained the mean [Fe/H] abundances for the four main components using the relation between [Fe/H] and (B-V)0 at fixed magnitude. One of the discoveries is that they identified for the first time a continuous and slanting RGB bump in the CMD of ω Cen, allowing them to confirm the existence of complex and multiple stellar populations with a wide spread in metallicity within ω Cen. They also determined the age-metallicity relation between the four distinct subpopulations.[9]
